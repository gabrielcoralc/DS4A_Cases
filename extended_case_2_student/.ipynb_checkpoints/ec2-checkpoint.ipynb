## Finalizing the cleaning for our study

Our data has improved a lot by now. However, since the number of indicators is still quite large, let us focus our study on the following indicators for now:

- **Total greenhouse gas emissions (kt of CO2 equivalent), EN.ATM.GHGT.KT.CE**: The total of greenhouse emissions includes CO2, Methane, Nitrous oxide, among other pollutant gases. Measured in kilotons.


- **CO2 emissions (kt), EN.ATM.CO2E.KT**: Carbon dioxide emissions are those stemming from the burning of fossil fuels and the manufacture of cement. They include carbon dioxide produced during consumption of solid, liquid, and gas fuels and gas flaring. 	


- **Methane emissions (kt of CO2 equivalent), EN.ATM.METH.KT.CE**: Methane emissions are those stemming from human activities such as agriculture and from industrial methane production.


- **Nitrous oxide emissions (kt of CO2 equivalent), EN.ATM.NOXE.KT.CE**: Nitrous oxide emissions are emissions from agricultural biomass burning, industrial activities, and livestock management.


- **Other greenhouse gas emissions, HFC, PFC and SF6 (kt of CO2 equivalent), EN.ATM.GHGO.KT.CE**: Other pollutant gases.


- **PM2.5 air pollution, mean annual exposure (micrograms per cubic meter), EN.ATM.PM25.MC.M3**: Population-weighted exposure to ambient PM2.5 pollution is defined as the average level of exposure of a nation's population to concentrations of suspended particles measuring less than 2.5 microns in aerodynamic diameter, which are capable of penetrating deep into the respiratory tract and causing severe health damage. Exposure is calculated by weighting mean annual concentrations of PM2.5 by population in both urban and rural areas.


- **PM2.5 air pollution, population exposed to levels exceeding WHO guideline value (% of total), EN.ATM.PM25.MC.ZS**: Percent of population exposed to ambient concentrations of PM2.5 that exceed the World Health Organization (WHO) guideline value.

#### 7.3
Compute a plot showing the behavior of each of the four main air pollutants for each of the main global regions in the ```Emissions_R_df``` data frame. The main regions are ```'Latin America & Caribbean', 'South Asia', 'Sub-Saharan Africa', 'Europe & Central Asia', 'Middle East & North Africa', 'East Asia & Pacific'``` and ```'North America'```. What conclusions can you make?


**Answer.**

````python
def Prov_Fun(x):
    MainR=['Latin America & Caribbean', 'South Asia', 'Sub-Saharan Africa', 'Europe & Central Asia', 'Middle East & North Africa', 'East Asia & Pacific','North America']
    bandera=0
    for MR in MainR:
        if x==MR:
            bandera=bandera+1
    if bandera==0:
        return False
    else:
        return True

Mainpll=["CO2","CH4","N20","Other"]
datap=Emissions_R_df.groupby(["Country Name","Indicator Code","Year"]).sum().reset_index()
datap["Year"]=pd.to_datetime(datap["Year"])
datap=datap[datap["Indicator Code"].apply(lambda x: any(z in x for z in Mainpll))]

datap=datap[datap["Country Name"].apply(Prov_Fun)]
datap["Country Name"].unique()
for Reg in datap["Country Name"].unique().tolist():
    plt.figure()
    fig, ax = pyplot.subplots(figsize=(30,10))
    for pll in Mainpll:
        datapp=datap[datap["Country Name"]==Reg]
        ax = sns.lineplot(x="Year", y="Indicator Value",data=datapp[datapp["Indicator Code"]==pll],label=pll)       
        ax.set_title('Air Pollution in '+ Reg)

plt.figure()
g = sns.FacetGrid(datap,col="Indicator Code",hue="Country Name",col_wrap=1,aspect= 5,) 
g = (g.map(sns.lineplot, "Year","Indicator Value").add_legend())
#The region with the highest pollution indicators is East Asia & Pacific, except for the Other indicator
````

#### 8.2

Are these five countries carrying out the burden of most of the emissions emitted over the years globally? Can we say that the rest of the world is making some effort to control their polluted gasses emissions over the years?

**Answer.**

````python
for I_Code in ["CO2","CH4","N20","Other"] :
    provdf=pd.DataFrame()
    for name, group in Emissions_C_df[Emissions_C_df["Indicator Code"]==I_Code].groupby("Year"):
        g=group.sort_values(by="Indicator Value",ascending=False).head(10).reset_index()[["Country Name","Year"]]
        provdf=pd.concat([provdf,g])
    MostContCountries=provdf["Country Name"].value_counts().reset_index().head(5)["index"].values
    Group_wout_m=Emissions_C_df[Emissions_C_df["Country Name"].apply(lambda x: x not in MostContCountries)]
    Group_wout_m=Group_wout_m[Group_wout_m["Indicator Code"]==I_Code].groupby("Year").sum().reset_index()
    Group_wout_m["Year"]=pd.to_datetime(Group_wout_m["Year"])
    plt.figure()
    fig, ax = pyplot.subplots(figsize=(30,10))
    for val in range(len(MostContCountries)):
        z=Emissions_C_df[Emissions_C_df["Country Name"]==MostContCountries[val]]
        z=z[z["Indicator Code"]==I_Code]
        z["Year"]=pd.to_datetime(z["Year"])
        ax=sns.lineplot(x="Year", y="Indicator Value",data=z,label=MostContCountries[val])
    ax2 = plt.twinx()
    #sns.lineplot(data=df.column2, color="b", ax=ax2)
    ax= sns.lineplot(x="Year", y="Indicator Value",data=Group_wout_m,label="The rest of the World", ax=ax2, color="y")
    ax.set_title("Emission of "+ I_Code +" (Top Five Countries & the rest of the world)")
    
````

## The health impacts of air pollution

### Exercise 9:

One of the main contributions of poor health from air pollution is particulate matter. In particular, very small particles (those with a size less than 2.5 micrometres ($\mu$m)) can enter and affect the respiratory system. The ```PM2.5``` indicator measures the average level of exposure of a nation's population to concentrations of these small particles. The ```PM2.5_WHO``` measures the  percentage of the population who are exposed to ambient concentrations of these particles that exceed some thresholds set by the World Health Organization (WHO). In particular, countries with a higher ```PM2.5_WHO``` indicator are more likely to suffer from bad health conditions.

#### 10.2

Use the indicators provided in Exercise 10.1 to give valuable information to the client. 

**Answer.**

````python
hh=WDI_ids[WDI_ids["Topic"].apply(lambda x: "Health: Mortality" in x)][["Series Code","Topic"]].reset_index().drop("index",axis=1)
keys=hh["Series Code"].unique().tolist()
Mortality_df=WDI_data[WDI_data["Indicator Code"].apply(lambda x: any(z in x for z in keys))]
Years=Mortality_df.columns.tolist()[4:-1]
Vars=Mortality_df.columns.tolist()[0:4]
Mortality_df=pd.melt(Mortality_df,id_vars=Vars,value_vars=Years).rename(columns={"variable":"Year","value":"Indicator Value"}).dropna()
Mortality_df["Indicator Name"].unique()

#Here we find three important indicators 
 #   'Mortality rate attributed to household and ambient air pollution, age-standardized (per 100,000 population)',
 #   'Mortality rate attributed to household and ambient air pollution, age-standardized, female (per 100,000 female population)',
 #   'Mortality rate attributed to household and ambient air pollution, age-standardized, male (per 100,000 male population)',
Mortality_df=Mortality_df[Mortality_df["Indicator Name"].apply(lambda x: "air pollution" in x )]
Mortality_df["Year"].unique()
Rename={'Mortality rate attributed to household and ambient air pollution, age-standardized (per 100,000 population)':"Mortality rate",
       'Mortality rate attributed to household and ambient air pollution, age-standardized, female (per 100,000 female population)':"Female Mortality rate",
        'Mortality rate attributed to household and ambient air pollution, age-standardized, male (per 100,000 male population)':"Male Mortality rate"}
Mortality_df["Indicator Name"]=Mortality_df["Indicator Name"].apply(lambda x : Rename[x])
Mortality_C_df=Mortality_df[Mortality_df["Country Code"].apply(lambda x: any(z in x for z in Country_Codes))]
Mortality_R_df=Mortality_df[Mortality_df["Country Code"].apply(lambda x: any(z in x for z in Region_Codes))]
#sns.distplot(Mortality_C_df[Mortality_C_df["Indicator Name"]=="Mortality rate"]["Indicator Value"])
g = sns.FacetGrid(Mortality_C_df, col="Indicator Name",col_wrap=1,aspect=5)
g = g.map(sns.distplot, "Indicator Value")
Mortality_rate=Mortality_C_df[Mortality_C_df["Indicator Name"]=="Mortality rate"].groupby("Country Name").mean().reset_index().rename(columns={"Indicator Value":"Mortality"})
Mortality_rate
````
